```python
import pandas as pd
import os

# Define the directory name and file name
directory_name = "customer data analysis"
csv_file_name = "dataset.csv"

# Construct the full path for the CSV file
csv_file_path = os.path.join(directory_name, csv_file_name)

# Read the CSV file into a DataFrame named 'raw'
try:
    raw = pd.read_csv(csv_file_path)

    # Display the DataFrame
    display(raw)

except FileNotFoundError:
    print(f"Error: The file '{csv_file_path}' was not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```

# Task
Analyze the dataset located at "/content/customer data analysis/Customer Data Analysis/dataset.xlsx" by converting it to a CSV, loading it into a DataFrame named `raw`, removing the 'Unnamed: 0' column, examining data types, converting the 'timestamp' column to datetime, checking for missing values, displaying summary statistics for numerical columns, and exploring unique values in categorical columns.

## Convert timestamp to datetime

### Subtask:
Convert the 'timestamp' column to a datetime data type.


**Reasoning**:
Convert the 'timestamp' column to datetime and display the data types to confirm the conversion.



```python
# Convert the 'timestamp' column to datetime objects
raw['timestamp'] = pd.to_datetime(raw['timestamp'])

# Display the data types of the DataFrame to confirm the conversion
print(raw.info())
```

## Check for missing values

### Subtask:
Identify and report any missing values in the DataFrame.


**Reasoning**:
Check for missing values in the DataFrame and display the count for each column.



```python
# Check for missing values and sum them for each column
missing_values = raw.isnull().sum()

# Display the sum of missing values for each column
print("Missing values per column:")
print(missing_values)
```

## Get summary statistics

### Subtask:
Display descriptive statistics for the numerical columns.


**Reasoning**:
Display descriptive statistics for the numerical columns.



```python
# Display descriptive statistics for numerical columns
display(raw.describe())
```

## Explore categorical columns

### Subtask:
Show the unique values and their counts for the categorical columns.


**Reasoning**:
Iterate through each column and print unique values and their counts for object type columns.



```python
for col in raw.columns:
    if raw[col].dtype == 'object':
        print(f"Unique values and counts for '{col}':")
        display(raw[col].value_counts())
        print("-" * 30)
```
